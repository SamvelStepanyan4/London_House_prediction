```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv('train.csv')
```

```python
df_test = pd.read_csv('test.csv')
```

```python
df_train.head(1)
```

```python
df_test.head(1)
```

```python
df_train.drop('ID',axis=1,inplace=True)
df_test.drop('ID',axis=1,inplace=True)
```

```python
df_train.isna().sum()*100/len(df_train)
```

```python
def fill_na(column_name,d_type):
    if d_type == 'train':
        elements = df_train[column_name].dropna().unique()
        proba = [df_train[column_name].value_counts()[i] / len(df_train[column_name].dropna()) for i in elements]
        df_train[column_name].fillna(np.random.choice(a=elements,p=proba),inplace=True)
    else:
        elements = df_test[column_name].dropna().unique()
        proba = [df_test[column_name].value_counts()[i] / len(df_test[column_name].dropna()) for i in elements]
        df_test[column_name].fillna(np.random.choice(a=elements,p=proba),inplace=True)
```

```python
for column in df_train.isna().sum()[df_train.isna().sum() > 0].index:
    fill_na(column,'train')
```

```python
for column in df_test.isna().sum()[df_test.isna().sum() > 0].index:
    fill_na(column,'test')
```

```python
df_test.isna().sum()*100/len(df_train)
```

```python
for i in df_test.columns:
    if df_train[i].dtype == object:
        if len(set(df_test[i]).difference(set(df_train[i]))) > 0:
            print(i)
            print(set(df_test[i]).difference(set(df_train[i])))
```

```python
df_train.drop('currentEnergyRating',axis=1,inplace=True)
df_test.drop('currentEnergyRating',axis=1,inplace=True)
```

```python
df_train.drop('sale_month',axis=1,inplace=True)
df_test.drop('sale_month',axis=1,inplace=True)
```

```python
df_train.drop('propertyType',axis=1,inplace=True)
df_test.drop('propertyType',axis=1,inplace=True)
```

```python
df_train.drop('latitude',axis=1,inplace=True)
df_test.drop('latitude',axis=1,inplace=True)
```

```python
df_train.drop('tenure',axis=1,inplace=True)
df_test.drop('tenure',axis=1,inplace=True)
```

```python
df_train.drop('fullAddress',axis=1,inplace=True)
df_test.drop('fullAddress',axis=1,inplace=True)
```

```python
df_train.drop('postcode',axis=1,inplace=True)
df_test.drop('postcode',axis=1,inplace=True)
```

```python
df_train['outcode'].value_counts()
```

```python
df_test['outcode'].replace(to_replace='IG8',value='SE18',inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in df_test.columns:
    if df_test[i].dtype == 'object':
        label = LabelEncoder()
        df_train[i] = label.fit_transform(df_train[i])
        df_test[i] = label.transform(df_test[i])
```

```python
df_train.corr()
```

# LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures()
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('scaler',scaler),('poly',poly),('model',model)])
```

```python
pipe
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'poly__degree' : [5],
}
```

```python
final_model = GridSearchCV(pipe,param_grid=params,cv=3,verbose=2)
```

```python
X_train = df_train.drop('price',axis=1)
X_test = df_test.copy()
y_train = df_train['price']
```

```python
final_model.fit(X_train,y_train)
```

```python
final_model.best_params_
```

```python
final_model.best_score_
```

```python
y_pred = final_model.predict(X_test)
```

```python
df_sub = pd.read_csv('sample_submission.csv')
```

```python
df_sub['price'] = y_pred
```

```python
df_sub.to_csv('sub.csv',index=False,index_label=False)
```
